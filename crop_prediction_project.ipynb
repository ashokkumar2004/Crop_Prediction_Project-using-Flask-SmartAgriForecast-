```python
# prompt: label encode the label column in dataset name df

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['label'] = le.fit_transform(df['label'])
df
```

```python
# prompt: split the input and  output the output column is label,remaining all are input

X = df.drop('label', axis=1)
y = df['label']
```

```python
# prompt: train test split using test size 30 %

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# prompt: build a svm model and find the accuracy

from sklearn import svm
from sklearn.metrics import accuracy_score
svc = svm.SVC()
model=svc.fit(X_train, y_train)
y_pred = model.predict(X_test)
print(accuracy_score(y_test, y_pred))
```

```python
# prompt: download the model in pickle format

import pickle

# Save the model to a pickle file
filename = 'crop_prediction_model svc.pkl'
pickle.dump(model, open(filename, 'wb'))

# Download the file to your local machine
from google.colab import files
files.download('crop_prediction_model svc.pkl')
```
